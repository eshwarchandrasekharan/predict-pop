```python
###for each buzz_id, get the external ID which was posted FIRST!!!
train = temp

train = train[~(train['content_id'] == -1)]
###
train_grouped = train.groupby(['content_id']).agg({'created_at':'min'})

train_grouped = train_grouped.reset_index()

train_grouped = train_grouped.rename(columns={'created_at':'first_post'})

df = pd.merge(train, train_grouped, how='left', on=['content_id'])

df = df[df['created_at'] == df['first_post']]
###
# train = temp

# temp = train[train['content_id'] == -1]
# df = pd.concat([df,temp])
train
```

```python
###get the previous page the video was posted on!

def get_previous(X):
#     print("Start")
    videoid = train[train['external_id'] == X]['content_id'].values[0]
    cross_posts = list(train[train['content_id'] == videoid].sort_values(['created_at'], ascending = True)['external_id']) 
    print(X, ":", videoid, ":", len(cross_posts))
#     print(cross_posts)
    index = cross_posts.index(X)
    if index == 0:
        return -1
#         return "-1"
    else:
        return cross_posts[index - 1]

df = train
df['previous_external_id'] = df['external_id'].apply(get_previous)
```

```python
###get cross-posting info for videos!!!! 
import math 

filepath = "/Users/eshwarchandrasekharan/Desktop/repo/predict-pop/code/"
# train = pd.read_csv('../code/jan_may_2017_links_cross_posting_info.csv')
train = pd.read_csv('stats_info_videos_jan_jun16_type_2_16.csv')

df = train.sort_values('created_at', ascending = False).drop_duplicates(subset=['content_id', 'external_id'], keep = 'last')

all_pages = list(df.display_page.unique())
print("Total pages considered: ", len(all_pages))

def get_cross_posted_pages(x):
        t = df[df.content_id == x].sort_values('created_at', ascending = 1)
        return t.display_page.unique()

df['cross_promotion_pages'] = df['content_id'].apply(get_cross_posted_pages)

def generate_cat_pages(x, y):
        if(y in x):
            return list(x).index(y) + 1
        else:
            return math.inf

###categorical features for pages where the link was shared
for pag in all_pages:
#     print("Getting for ", pag)
    df[pag] = df['cross_promotion_pages'].apply(generate_cat_pages, y = pag)

# df.to_csv('jan_may_2017_videos_cross_posting_info.csv', index = False)
```
